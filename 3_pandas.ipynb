- Объекты класса DataFrame

```python
area_d = {
    "Нерюнгри": 99,
    "Вилюйск": 15,
    "Якутск": 122
}

area_s = pd.Series(area_d)
area_s
```

```python
ykt_df = pd.DataFrame({"population": pop_s, "area": area_s})
ykt_df
```

```python
# данные из столбца 
ykt_df["area"]
```

```python
# данные из строки
# комбинация индексации по строкам и среза по столбцам
ykt_df.loc["Якутск", :]
```

```python
# индексация по строкам и по столбцам 
ykt_df.loc["Якутск", "population"]
```

```python
direct_d = {
    "Нерюнгри": "s",
    "Вилюйск": "w",
    "Якутск": "c"
}

ykt_df["direction"] = direct_d
ykt_df
```

```python
# срез по строкам и прихотливая индексация по столбцам
ykt_df.loc["Нерюнгри":"Якутск", ["direction"]]
```

```python
# маскирование (фильтрация данных)
mask = ykt_df["population"] > 40000
mask
```

```python
# комбинация маскирования по строкам и срез по столбцам
ykt_df.loc[mask, :]
```
