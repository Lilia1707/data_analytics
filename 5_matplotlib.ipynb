```python
import numpy as np
import matplotlib.pyplot as plt
```

 - Линейные графики

```python
# 1-мерный массив
x_arr = np.arange(-10, 10)
x_arr
```

```python
y_arr = x_arr ** 2
y_arr
```

```python
# визуализация
plt.figure(figsize=(10, 8))

plt.plot(x_arr, y_arr)

plt.show()
```

```python
# совмещение нескольких графиков в одном окне

y_arr_1 = x_arr  * 2
y_arr_2 = x_arr ** 3

plt.plot(x_arr, y_arr_1)
plt.plot(x_arr, y_arr_2)

plt.show()
```

```python
# 1-мерный массив из 100 значений
# в диапазоне от -10 до 10
x_arr_2 = np.linspace(-10, 10, 100)
x_arr_2 
```

```python
# функция синуса 
y_1 = np.sin(x_arr_2)

# функция гиперболического тангенса
y_2 = np.tanh(x_arr_2)

# функция сигмоиды
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_3 = sigmoid(x_arr_2)

plt.plot(x_arr_2, y_1)
plt.plot(x_arr_2, y_2)
plt.plot(x_arr_2, y_3)

plt.show()
```

```python
# цвета линий

y_1 = np.sin(x_arr_2)
y_2 = np.sin(x_arr_2 + 1)
y_3 = np.sin(x_arr_2 + 2)

# по названию цвета
plt.plot(x_arr_2, y_1, c="green")
# по букве названия цвета из форматов RGB или CMYK
plt.plot(x_arr_2, y_2, c="m")
# по 16-ричной кодировке RGB или RGBA
plt.plot(x_arr_2, y_3, c="#dff842")

plt.show()
```

```python
# типы линий

# штриховая
plt.plot(x_arr_2, y_1, linestyle="dashed")
# пунктирная линия
plt.plot(x_arr_2, y_2, ls="dotted")
# штрих-пунктирная линия
plt.plot(x_arr_2, y_3, ls="dashdot", c="pink", lw=3)

plt.show()
```

```python
# заголовок, легенда графика, подписи осей 

plt.figure(figsize=(10,7))

y_1 = np.sin(x_arr_2)
y_2 = np.cos(x_arr_2)

plt.plot(x_arr_2, y_1)
plt.plot(x_arr_2, y_2)

# заголовок
plt.title("Графики синуса и косинуса")

# легенда графиков
plt.legend(["синус", "косинус"])

# подписи осей
plt.xlabel("ось Х")
plt.ylabel("ось У")
plt.show()
```
